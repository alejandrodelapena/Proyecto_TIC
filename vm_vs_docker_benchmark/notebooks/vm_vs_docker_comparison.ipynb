```python

import pandas as pd
import matplotlib.pyplot as plt
import statistics
import os

def calcular_media(columna):
    return statistics.mean(columna)

# Archivos de entrada
vm_file = "../results/vm_metrics/metrics_vm.csv"
docker_file = "../results/docker_metrics/metrics_docker.csv"

if not os.path.exists(vm_file) or not os.path.exists(docker_file):
    raise FileNotFoundError(" Falta uno de los archivos de mtricas.")

# Cargar datos
vm = pd.read_csv(vm_file)
docker = pd.read_csv(docker_file)

# Calcular medias
vm_cpu_media = calcular_media(vm["cpu_percent"])
docker_cpu_media = calcular_media(docker["cpu_percent"])

vm_ram_media = calcular_media(vm["mem_percent"])
docker_ram_media = calcular_media(docker["mem_percent"])

vm_rx_media = calcular_media(vm["net_rx_kbps"])
docker_rx_media = calcular_media(docker["net_rx_kbps"])

vm_tx_media = calcular_media(vm["net_tx_kbps"])
docker_tx_media = calcular_media(docker["net_tx_kbps"])

# Grfico 1: Uso de CPU
plt.figure(figsize=(10, 4))
plt.plot(vm["cpu_percent"], label="VM CPU")
plt.plot(docker["cpu_percent"], label="Docker CPU")
plt.axhline(vm_cpu_media, color='blue', linestyle='--', label=f"Media VM CPU ({vm_cpu_media:.2f}%)")
plt.axhline(docker_cpu_media, color='orange', linestyle='--', label=f"Media Docker CPU ({docker_cpu_media:.2f}%)")
plt.legend()
plt.title("Uso de CPU: VM vs Docker")
plt.xlabel("Muestras")
plt.ylabel("% CPU")
plt.grid(True)
plt.tight_layout()
plt.show()

# Grfico 2: Uso de RAM
plt.figure(figsize=(10, 4))
plt.plot(vm["mem_percent"], label="VM RAM")
plt.plot(docker["mem_percent"], label="Docker RAM")
plt.axhline(vm_ram_media, color='blue', linestyle='--', label=f"Media VM RAM ({vm_ram_media:.2f}%)")
plt.axhline(docker_ram_media, color='orange', linestyle='--', label=f"Media Docker RAM ({docker_ram_media:.2f}%)")
plt.legend()
plt.title("Uso de RAM: VM vs Docker")
plt.xlabel("Muestras")
plt.ylabel("% RAM")
plt.grid(True)
plt.tight_layout()
plt.show()

# Grfico 3: Trfico de Red RX
plt.figure(figsize=(10, 4))
plt.plot(vm["net_rx_kbps"], label="VM RX")
plt.plot(docker["net_rx_kbps"], label="Docker RX")
plt.axhline(vm_rx_media, color='blue', linestyle='--', label=f"Media VM RX ({vm_rx_media:.2f} KB/s)")
plt.axhline(docker_rx_media, color='orange', linestyle='--', label=f"Media Docker RX ({docker_rx_media:.2f} KB/s)")
plt.legend()
plt.title("Trfico de Red RX: VM vs Docker")
plt.xlabel("Muestras")
plt.ylabel("KB/s")
plt.grid(True)
plt.tight_layout()
plt.show()

# Grfico 4: Trfico de Red TX
plt.figure(figsize=(10, 4))
plt.plot(vm["net_tx_kbps"], label="VM TX")
plt.plot(docker["net_tx_kbps"], label="Docker TX")
plt.axhline(vm_tx_media, color='blue', linestyle='--', label=f"Media VM TX ({vm_tx_media:.2f} KB/s)")
plt.axhline(docker_tx_media, color='orange', linestyle='--', label=f"Media Docker TX ({docker_tx_media:.2f} KB/s)")
plt.legend()
plt.title("Trfico de Red TX: VM vs Docker")
plt.xlabel("Muestras")
plt.ylabel("KB/s")
plt.grid(True)
plt.tight_layout()
plt.show()




# Calcular medias de latencia
vm_latency_media = calcular_media(vm["latency_ms"])
docker_latency_media = calcular_media(docker["latency_ms"])

# Grfico 5: Latencia de red
plt.figure(figsize=(10, 4))
plt.plot(vm["latency_ms"], label="VM Latencia")
plt.plot(docker["latency_ms"], label="Docker Latencia")
plt.axhline(vm_latency_media, color='blue', linestyle='--', label=f"Media VM ({vm_latency_media:.2f} ms)")
plt.axhline(docker_latency_media, color='orange', linestyle='--', label=f"Media Docker ({docker_latency_media:.2f} ms)")
plt.legend()
plt.title("Latencia de Red: VM vs Docker")
plt.xlabel("Muestras")
plt.ylabel("Milisegundos (ms)")
plt.grid(True)
plt.tight_layout()
plt.show()

```
